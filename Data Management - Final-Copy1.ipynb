```python
# Identify the columns to check
top_columns = [f"Top{i}" for i in range(1, 26)]

matches = []

for ticker, keywords in tickers_keywords.items():
    # Build regex pattern for keywords of this ticker
    pattern = "|".join(rf"\b{k}\b" for k in keywords)
    
    # Iterate through the 25 columns
    for col in top_columns:
        # Boolean mask where keywords are found (case-insensitive)
        mask = news_df[col].str.contains(pattern, case=False, na=False)
        
        # Extract matching rows, but keep only Date, the matched text, and the ticker
        temp_df = news_df.loc[mask, ["Date", col]].copy()
        temp_df.rename(columns={col: "text"}, inplace=True)
        temp_df["ticker"] = ticker
        
        matches.append(temp_df)

# Combine all matched results
result_df = pd.concat(matches, ignore_index=True)
```

```python
result_df.rename(columns={'ticker': 'Company'}, inplace=True)
```
